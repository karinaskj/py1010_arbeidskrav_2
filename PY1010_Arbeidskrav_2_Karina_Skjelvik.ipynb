#### Oppgave 4 a b og c)

```python
# Oppgave 4a - Opprette en dictionary med fire land samt hovedstad og antall innbyggere i hovedstaden

data = {                                     
    "Norge": {
        "hovedstad": "Oslo",
        "innbyggere_hovedstad": 0.634
        },
    
    "England": {
        "hovedstad": "London",
        "innbyggere_hovedstad": 8.982
        },
    "Frankrike": {
        "hovedstad": "Paris",
        "innbyggere_hovedstad": 2.161
        },
    "Italia": {
        "hovedstad": "Roma",
        "innbyggere_hovedstad": 2.873
        }
    }

# Oppgave 4b - Be brukeren om å skrive inn et land

land = input("Skriv inn et land:")

hovedstad = data[land]["hovedstad"]
innbyggere_hovedstad = data[land]["innbyggere_hovedstad"]

print(hovedstad, "er hovedstaden i", land, "og det er", innbyggere_hovedstad, "mill. innbyggere i", hovedstad)

# Oppgave 4c - Legg til et nyttland i dictionary ved å be bruker om å legge inn input

nytt_land = input("Skriv inn et nytt land som kan legges i dictionary:")
nytt_land_hovedstad = input("Hva er hovedstaden?")
nytt_land_innbyggere_hovedstad = input("Hvor mange millioner innbyggere er det i hovedstaden?")

data[nytt_land] = {
        "hovedstad": nytt_land_hovedstad,
        "innbyggere_hovedstad": nytt_land_innbyggere_hovedstad
    }

print(nytt_land, "er lagt til i dictionary")


# Print ut oppdatert dictionary for å sjekke at nytt land er lagt inn
print(data)

```

```python
import math

a = 3   # Diameter av sirkelen
b = 4   #Side av trekanten

areal_sirkel = math.pi * (a/2)**2
omkrets_sirkel = 2 * math.pi * a/2

c = math.sqrt(a**2 + b**2)    # Hypotenus av den rettvinklede trekanten
areal_trekant = a * b / 2
omkrets_trekant = a + b + c


omkrets_figur = b + c + ( omkrets_sirkel / 2)

areal_figur = areal_trekant + (areal_sirkel / 2 )

print("For a = 3cm og b = 4cm er arealet og ytre omkrets av figuren hhv.", round(areal_figur,2), "cm og", round(omkrets_figur,2), "cm")

```
